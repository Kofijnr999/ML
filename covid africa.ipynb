```python
Data=dict(type='choropleth',
         locations=covid['Country'],
         locationmode='country names',
         colorscale='Viridis',
         z=covid['Total Cases'],
         text=covid['Total Deaths'],
         colorbar={'title':'COVID 19 STATS IN AFRICA'})


Data
```

```python
layout=dict(title='COVID 19 STATS IN AFRICA',
           geo=dict(showframe=False,
                   projection={'type':'mercator'}),
           scope='africa',
           lakecolor='rgb(0,0,255)')
layout
```

```python
choropleth=go.Figure(data=[Data],layout=layout)

iplot(choropleth)
```
